## rename video and srt files according to excel input file

```python
import pandas as pd
import os

# Specify the path to your Excel file
excel_file_path = r"C:\Users\Sean\OneDrive - Virginia Tech\Documents\SGC_Teaching_MSE_2034_D2L\vscode_brightspace_development\brightspace\materials_science_renderings\assets\media\file_list.xlsx"

def rename_files(excel_path):
    try:
        data = pd.read_excel(excel_path)
    except Exception as e:
        print(f"Failed to read the Excel file: {e}")
        return

    for index, row in data.iterrows():
        directory_path = row['directory path']
        video_file_name = row['video file name']
        srt_file_name = row['srt file name']
        new_name = row['rename']

        if pd.isna(new_name):
            print(f"Missing new name in row {index}: skipping this row.")
            continue

        # Process video file if it exists
        if not pd.isna(video_file_name):
            video_path = os.path.join(directory_path, video_file_name)
            video_extension = os.path.splitext(video_file_name)[1]
            new_video_path = os.path.join(directory_path, new_name + video_extension)
            if os.path.exists(video_path):
                os.rename(video_path, new_video_path)
                print(f"Renamed '{video_file_name}' to '{new_name + video_extension}'")
            else:
                print(f"Video file not found: {video_path}")

        # Process subtitle file if it exists
        if not pd.isna(srt_file_name):
            srt_path = os.path.join(directory_path, srt_file_name)
            srt_extension = os.path.splitext(srt_file_name)[1]
            new_srt_path = os.path.join(directory_path, new_name + srt_extension)
            if os.path.exists(srt_path):
                os.rename(srt_path, new_srt_path)
                print(f"Renamed '{srt_file_name}' to '{new_name + srt_extension}'")
            else:
                print(f"SRT file not found: {srt_path}")

# Specify the path to your Excel file

rename_files(excel_file_path)

```
